```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hospital = pd.read_csv("data/df_betim.csv")
df_hospital.rename(columns={"DT_INTER": 'Date'}, inplace=True)
```

```python
df_hospital['Date'] = pd.to_datetime(df_hospital['Date'], format='%Y%m%d')
df_hospital.head()
```

```python
df_hospital.isnull().sum()
```

```python
df_days = pd.read_csv("data/days_weather_pampulha.csv") 
df_days['Date'] = pd.to_datetime(df_days['Date'])
df_days.head()
```

```python
df_day_hospital = pd.merge(df_days, df_hospital, on='Date', how='left')
```

```python
df_day_hospital.head(20)
```

```python
df_day_hospital.describe()
```

```python
df_day_hospital.isnull().sum()
```

```python
# Days with no hospital admi
df_day_hospital['NUMBER_INT'] = df_day_hospital['NUMBER_INT'].fillna(0.0)
```

```python
df_day_hospital.isnull().sum()
```

```python
df_day_hospital.tail()
```

```python
df_bairro = pd.read_csv("data/betim_pivot_bairro.csv")
df_bairro.head()
```

```python
# df_bairro['Data'] = pd.to_datetime(df_bairro['Data'], format='%d/%m/%Y').dt.strftime('%Y%m%d').astype(int)
df_bairro['Data'] = pd.to_datetime(df_bairro['Data'], format='%d/%m/%Y')

df_bairro.head(20)
```

```python
df_bairro.rename(columns={'Data': 'Date', 'CO': 'EST1_CO', 'MP10': 'EST1_MP10', 'O3': 'EST1_03', 'MP2.5': 'EST1_MP2.5', 'SO2': 'EST1_SO2'}, inplace=True)
```

```python
df_day_hospital_betim = pd.merge(df_day_hospital, df_bairro, on='Date', how='left')
```

```python
df_day_hospital_betim.head()
```

```python
df_day_hospital_betim.tail()
```

```python
#export final dataset
df_day_hospital_betim.to_csv('df_final_betim_bairro.csv', index=False)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_day_hospital_betim['NUMBER_INT'], bins=20)
plt.title('Distribution of Hospitalizations')
plt.xlabel('Number of Admissions')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='temp_max', y='NUMBER_INT', data=df_day_hospital_betim)
plt.title('Temperature (Max) vs Hospitalizations')
plt.xlabel('Max Temperature (C)')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.show()
```

```python
selected_columns = [
    'Date', 'DayOfWeekNum', 'IsWeekend', 
    'temp_max', 'temp_min', 'humidity_max', 
    'humidity_min', 'NUMBER_INT', 
    'EST1_CO', 'EST1_MP10', 'EST1_03', 'EST1_MP2.5', 'EST1_SO2'
]

df_selected_num = df_day_hospital_betim[selected_columns]

correlation_matrix = df_selected_num.corr()
plt.figure(figsize=(16, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_selected_num.corr()['NUMBER_INT'].sort_values(ascending=False)
```

```python
weekend_vs_weekday = df_day_hospital_betim.groupby('IsWeekend')['NUMBER_INT'].mean()
plt.bar(['Weekday', 'Weekend'], weekend_vs_weekday)
plt.title('Average Hospitalizations: Weekday vs Weekend')
plt.ylabel('Average Admissions')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_day_hospital_betim['NUMBER_INT'])
plt.title('Boxplot - Number of Admissions')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='EST1_MP10', y='NUMBER_INT', data=df_day_hospital_betim, alpha=0.5)
plt.title('Hospitalizations vs PM10 Levels')
plt.xlabel('PM10 Levels')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='temp_min', y='NUMBER_INT', data=df_day_hospital_betim, alpha=0.5, color='orange')
plt.title('Hospitalizations vs Minimum Temperature')
plt.xlabel('Minimum Temperature (C)')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(25, 8))

df_day_hospital_betim_copy = df_day_hospital_betim.copy()

df_day_hospital_betim_copy['Date'] = pd.to_datetime(df_day_hospital_betim_copy['Date'], format='%Y%m%d')

features_plot = [
    'temp_max', 'temp_min', 'NUMBER_INT',
    'EST1_CO', 'EST1_MP10', 'EST1_03', 'EST1_MP2.5', 'EST1_SO2'
]



for feature in features_plot:
    plt.plot(df_day_hospital_betim_copy['Date'], 
             df_day_hospital_betim_copy[feature], 
             label=feature)

plt.title('Trends of Key Features Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25, 8))

features_plot_with_day = [
    'DayOfWeekNum',
    'temp_max', 
    'temp_min', 
    'NUMBER_INT'
]



for feature in features_plot_with_day:
    plt.plot(df_day_hospital_betim_copy['Date'], 
             df_day_hospital_betim_copy[feature], 
             label=feature)

plt.title('Trends of Key Features Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weekdays_avg = df_day_hospital_betim.groupby('DayOfWeekNum')['NUMBER_INT'].mean()

day_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(8, 5))
plt.bar(day_labels, weekdays_avg)
plt.title('Average Hospitalizations by Day of the Week')
plt.ylabel('Average Admissions')
plt.show()
```
