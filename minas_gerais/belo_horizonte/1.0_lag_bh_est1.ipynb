```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("df_final_bh_amazonas.csv")
```

```python
df.tail(35)
```

```python
df.count()
```

# LAG - CO

```python
selected_cols_co = [
    'Date',
    'NUMBER_INT',
    'EST1_CO'
]

df_selected_co = df[selected_cols_co].copy()
```

```python
df_selected_co.count()
```

```python
# df_selected_co = df_selected_co.dropna()
```

```python

df_selected_co['EST1_CO_lag1'] = df_selected_co['EST1_CO'].shift(1)
df_selected_co['EST1_CO_lag2'] = df_selected_co['EST1_CO'].shift(2)
df_selected_co['EST1_CO_lag3'] = df_selected_co['EST1_CO'].shift(3)
df_selected_co['EST1_CO_lag4'] = df_selected_co['EST1_CO'].shift(4)
df_selected_co['EST1_CO_lag5'] = df_selected_co['EST1_CO'].shift(5)
df_selected_co['EST1_CO_lag6'] = df_selected_co['EST1_CO'].shift(6)
df_selected_co['EST1_CO_lag7'] = df_selected_co['EST1_CO'].shift(7)

# df_selected_co = df_selected_co.reset_index(drop=True)
```

```python
# df_selected_co = df_selected_co.dropna()
```

```python
df_selected_co.count()
```

```python
correlation_matrix = df_selected_co.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
