```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read data

```python
df = pd.read_csv("df_final_bh_amazonas.csv")
```

```python
df.head()
```

```python
df.count()
```

# LAG - CO

```python
selected_cols_co = [
    'Date',
    'NUMBER_INT',
    'CO'
]

df_selected_co = df[selected_cols_co].copy()
```

```python
df_selected_co.count()
```

```python

df_selected_co['CO_lag1'] = df_selected_co['CO'].shift(1)
df_selected_co['CO_lag2'] = df_selected_co['CO'].shift(2)
df_selected_co['CO_lag3'] = df_selected_co['CO'].shift(3)
df_selected_co['CO_lag4'] = df_selected_co['CO'].shift(4)
df_selected_co['CO_lag5'] = df_selected_co['CO'].shift(5)
df_selected_co['CO_lag6'] = df_selected_co['CO'].shift(6)
df_selected_co['CO_lag7'] = df_selected_co['CO'].shift(7)

# df_selected_co = df_selected_co.reset_index(drop=True)
```

```python
# df_selected_co = df_selected_co.dropna()
```

```python
df_selected_co.count()
```

```python
correlation_matrix = df_selected_co.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
