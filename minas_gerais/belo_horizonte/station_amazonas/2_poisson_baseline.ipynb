# Poisson distribution

https://omarfsosa.github.io/poisson_regression_in_python

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# X = df_final.drop(columns=['Date', 'NUMBER_INT'])
# y = df_final['NUMBER_INT']

# X_train, X_val, y_train, y_val = train_test_split(
#     X, y, test_size=0.2, random_state=1
# )






# # Ensure 'Date' is datetime type
df_final['Date'] = pd.to_datetime(df_final['Date'])

# Sort by date just in case
df_final = df_final.sort_values(by='Date')

# Define training and testing periods
train_end_date = df_final['Date'].dt.year.min() + 4  # first 4 years
train_mask = df_final['Date'].dt.year < train_end_date
test_mask = df_final['Date'].dt.year == train_end_date  # 5th year

# Split features and target
X_train = df_final.loc[train_mask].drop(columns=['Date', 'NUMBER_INT'])
y_train = df_final.loc[train_mask, 'NUMBER_INT']

X_val = df_final.loc[test_mask].drop(columns=['Date', 'NUMBER_INT'])
y_val = df_final.loc[test_mask, 'NUMBER_INT']





# Add a constant for the intercept
X_train_sm = sm.add_constant(X_train)
X_val_sm = sm.add_constant(X_val)



# Fit Poisson Regression model
poisson_model = sm.GLM(y_train, X_train_sm, family=sm.families.Poisson())
poisson_results = poisson_model.fit()

y_pred = poisson_results.predict(X_val_sm)


print(poisson_results.summary())
```

```python
mse_poisson = mean_squared_error(y_val, y_pred)
rmse_poisson = np.sqrt(mse_poisson)
r2_poisson = r2_score(y_val, y_pred)

print(f"Mean Squared Error (MSE): {mse_poisson:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_poisson:.2f}")
print(f"R-squared (R²): {r2_poisson:.2f}")
```
