```python
import pandas as pd

df = pd.read_csv('air_q_b_horizonte_amazonas.csv')
# https://medium.com/@filip.sekan/pivot-vs-pivot-table-methods-in-pandas-definitions-examples-differences-dcc9e14e667f
```

```python
df_filtered_SO2 = df[df['Poluente'] == 'SO2'].copy()
```

```python
df_filtered_SO2.head()
```

```python
df['IQAR'].unique()
```

```python
poluentes = ['CO', 'MP10', 'O3', 'MP2.5', 'SO2']

df_filtrado = df[df['Poluente'].isin(poluentes)]

df_pivot = df_filtrado.pivot(index='Data', columns='Poluente', values='NumeroIQAR')

df_pivot = df_pivot.reindex(columns=poluentes)

df_pivot.reset_index(inplace=True)
```

```python
df_pivot.head()
```

```python
df_pivot['Data'] = pd.to_datetime(df_pivot['Data'], format='%d/%m/%Y')
df_pivot.sort_values(by='Data', inplace=True)
```

```python
df_pivot_filtered = df_pivot[df_pivot['SO2'].notna()]
```
