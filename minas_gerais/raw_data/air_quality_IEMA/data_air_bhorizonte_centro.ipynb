```python
import pandas as pd

df = pd.read_csv('air_q_b_horizonte_centro.csv')
# https://medium.com/@filip.sekan/pivot-vs-pivot-table-methods-in-pandas-definitions-examples-differences-dcc9e14e667f
```

```python
df.head()
```

```python
poluentes = ['CO', 'MP10', 'O3', 'MP2.5', 'SO2']

df_filtrado = df[df['Poluente'].isin(poluentes)]

df_pivot = df_filtrado.pivot(index='Data', columns='Poluente', values='NumeroIQAR')

df_pivot = df_pivot.reindex(columns=poluentes)

df_pivot.reset_index(inplace=True)
```

```python
df_pivot.head()
```

```python
df_pivot.to_csv('b_horizonte_pivot_centro.csv', index=False, encoding='utf-8-sig')
```
