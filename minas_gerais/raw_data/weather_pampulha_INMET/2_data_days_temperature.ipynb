# Create dataset Days + weather (Pampulha Belo Horizonte 2015 - 2019)

```python
import pandas as pd
```

```python
## DF days
```

```python

date_range = pd.date_range(start='2015-01-01', end='2019-12-31', freq='D')

df = pd.DataFrame({
    'Date': date_range.strftime('%Y%m%d'),
    'DayOfWeek': date_range.strftime('%a'),
    'DayOfWeekNum': date_range.weekday,  # Monday=0, Sunday=6
    'IsWeekend': date_range.weekday.map(lambda x: 1 if x >= 5 else 0)
})


df.tail()
# df.to_csv('complete_date_dataset.csv', index=False)
```

## Add Weather
