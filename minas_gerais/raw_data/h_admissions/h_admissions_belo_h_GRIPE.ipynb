# Filter for GRIPE

```python
filtered_df = df_belo_horizonte[df_belo_horizonte['def_procedimento_realizado'] == 'TRATAMENTO DE PNEUMONIAS OU INFLUENZA (GRIPE)']
```

```python
filtered_df['def_procedimento_realizado'].value_counts(normalize=True)
```

# Group by day

```python
result_df = filtered_df.groupby('DT_INTER').size().reset_index(name='NUMBER_INT')

result_df.head(25)
```

```python
# Save
result_df.to_csv('df_b_horizonte_day_GRIPE.csv', index=False)
```

```python
result_df.describe()
```

```python
result_df['DT_INTER'] = pd.to_datetime(result_df['DT_INTER'], format='%Y%m%d')
```

```python
result_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.plot(result_df['DT_INTER'], result_df['NUMBER_INT'], marker='o')
plt.title('Hospitalizations Over Time')
plt.xlabel('Date of Admission (DT_INTER)')
plt.ylabel('Number of Admissions')
plt.show()
```
